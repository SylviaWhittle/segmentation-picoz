```python
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt

base_dir = Path("../")
```

```python
batch_x = np.load(base_dir / "batch_x.npy")
batch_y = np.load(base_dir / "batch_y.npy")

print(batch_x.shape)
print(batch_y.shape)

for sample in range(batch_y.shape[0]):
    for sample_class in range(batch_y.shape[3]):
        plt.subplot(1, batch_y.shape[3], sample_class + 1)
        plt.imshow(batch_y[sample, :, :, sample_class], cmap="gray")
        plt.title(f"Class {sample_class}")
    plt.show()
```
